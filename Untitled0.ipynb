**MISSING DATA HANDLING**

```python
df = pd.read_csv('AB_NYC_2019.csv')
print(df.isnull().sum())
```

**DUPLICATE REMOVAL**

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
df.isnull().sum()
```

**STANDARDIZATION**

```python
feature_columns = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

scaler = StandardScaler()
df[feature_columns] = scaler.fit_transform(df[feature_columns])
```

**OUTLIER DETECTION**

```python
numeric_columns = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

```

```python
z_scores = stats.zscore(df[numeric_columns])
threshold = 3
outliers = (abs(z_scores) > threshold)
```

```python
df_no_outliers = df[outliers.any(axis=1)]
```
